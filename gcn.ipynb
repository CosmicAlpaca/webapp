```python
# -*- coding: utf-8 -*-
"""
Mã nguồn hoàn chỉnh để huấn luyện mô hình bệnh nghề nghiệp
kết hợp dữ liệu bảng và dữ liệu hình ảnh.
Kiến trúc: Mô hình Đa Đầu Vào (Multi-Input)
Thư viện: PyTorch & PyTorch Geometric (PyG)
*** PHIÊN BẢN CUỐI CÙNG: Đã sửa lỗi KeyError ***
"""

import os
import gc
import pickle
import numpy as np
import pandas as pd
from PIL import Image
from tqdm.notebook import tqdm

import torch
import torch.nn.functional as F
from torch.nn import Linear, BatchNorm1d, Module, Sequential, ReLU, Dropout, BCEWithLogitsLoss
from torch_geometric.data import Data, Dataset, DataLoader
from torch_geometric.nn import GCNConv, global_mean_pool

import torchvision.transforms as T
from torchvision.models import mobilenet_v2, MobileNet_V2_Weights
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, MinMaxScaler
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score, f1_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns


# BƯỚC 2: THIẾT LẬP MÔI TRƯỜNG VÀ ĐƯỜNG DẪN
# -------------------------------------------------
from google.colab import drive
drive.mount('/content/drive')

BASE_DIR = "/content/"
EXCEL_FILE_PATH = os.path.join(BASE_DIR, "Main_data_fixed_Not_Encode_Mapping_New.xlsx")
IMAGE_DIR = os.path.join(BASE_DIR, "sillicosis")
OUTPUT_DIR = os.path.join(BASE_DIR, "output_model_pytorch_final")

os.makedirs(OUTPUT_DIR, exist_ok=True)
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Sử dụng thiết bị: {DEVICE}")

# --- Các hằng số ---
IMAGE_VECTOR_DIM = 1280
TARGET_COLUMN = 'bnn'
MODEL_FILE_NAME = "occupational_disease_pyg.pth"
COLUMNS_FILE_NAME = "COLUMNS.pkl"
ENCODER_FILE_NAME = "ORDINAL_ENCODER.pkl"
SCALER_FILE_NAME = "MIN_MAX_SCALER.pkl"


# BƯỚC 3: TRÍCH XUẤT ĐẶC TRƯNG ẢNH
# -------------------------------------------------
print("Bắt đầu tải mô hình trích xuất đặc trưng ảnh...")
weights = MobileNet_V2_Weights.IMAGENET1K_V2
image_feature_extractor = mobilenet_v2(weights=weights)
image_feature_extractor.classifier = torch.nn.Identity()
image_feature_extractor.to(DEVICE)
image_feature_extractor.eval()
image_transforms = weights.transforms()

def extract_image_features(image_path):
    if not image_path or not os.path.exists(image_path):
        return np.zeros(IMAGE_VECTOR_DIM)
    try:
        img = Image.open(image_path).convert('RGB')
        img_tensor = image_transforms(img).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            features = image_feature_extractor(img_tensor)
        return features.cpu().numpy().flatten()
    except Exception as e:
        return np.zeros(IMAGE_VECTOR_DIM)

# BƯỚC 4: XÂY DỰNG DATASET TÙY CHỈNH
# -------------------------------------------------
class OccupationalDiseaseDataset(Dataset):
    def __init__(self, root, tabular_df, image_features_array, targets):
        super(OccupationalDiseaseDataset, self).__init__(root)
        self.tabular_df, self.image_features_array, self.targets = tabular_df, image_features_array, targets
        self.columns = tabular_df.columns.tolist()
        self._build_graphs()

    def len(self):
        return len(self.graphs)

    def get(self, idx):
        graph_data = self.graphs[idx]
        image_features = torch.tensor(self.image_features_array[idx], dtype=torch.float)
        target = torch.tensor([self.targets[idx]], dtype=torch.float)
        return graph_data, image_features, target

    def _build_graphs(self):
        self.graphs = []
        relation_ship_groups = [['cviec1','tgian1'],['cviec2','tgian2'],['tuoinghe','A9a','A9b','nampx'],['hutthuoc','slthuoc'],['tiensuhh','benhhh'],['ho','tdho','tsho']]
        for _, row in self.tabular_df.iterrows():
            node_features = torch.tensor(row.values, dtype=torch.float).view(-1, 1)
            source_nodes, target_nodes = [], []
            num_nodes = len(self.columns)
            for i in range(num_nodes):
                for j in range(i + 1, num_nodes):
                    source_nodes.extend([i, j]); target_nodes.extend([j, i])
            for group in relation_ship_groups:
                node_indices = [self.columns.index(col) for col in group if col in self.columns]
                if len(node_indices) > 1:
                    first_node = node_indices[0]
                    for other_node in node_indices[1:]:
                        source_nodes.extend([first_node, other_node]); target_nodes.extend([other_node, first_node])
            edge_index = torch.tensor([source_nodes, target_nodes], dtype=torch.long)
            self.graphs.append(Data(x=node_features, edge_index=edge_index))

# BƯỚC 5: ĐỊNH NGHĨA KIẾN TRÚC MÔ HÌNH
# -------------------------------------------------
class MultiInputGCN(Module):
    def __init__(self, num_node_features, graph_embedding_dim=64, image_feature_dim=IMAGE_VECTOR_DIM):
        super(MultiInputGCN, self).__init__()
        self.gcn1 = GCNConv(num_node_features, 128)
        self.gcn2 = GCNConv(128, graph_embedding_dim)
        self.image_mlp = Sequential(Linear(image_feature_dim, 256), ReLU(), Dropout(0.5), Linear(256, graph_embedding_dim))
        self.classifier = Sequential(Linear(graph_embedding_dim * 2, 128), ReLU(), BatchNorm1d(128), Dropout(0.5), Linear(128, 1))

    def forward(self, graph_data, image_features):
        x, edge_index, batch = graph_data.x, graph_data.edge_index, graph_data.batch
        x = F.relu(self.gcn1(x, edge_index))
        x = self.gcn2(x, edge_index)
        graph_embedding = global_mean_pool(x, batch)
        image_embedding = self.image_mlp(image_features)
        combined_embedding = torch.cat([graph_embedding, image_embedding], dim=1)
        return self.classifier(combined_embedding)

# BƯỚC 6: TIẾN TRÌNH HUẤN LUYỆN VÀ ĐÁNH GIÁ
# -------------------------------------------------
def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss, total_correct, total_samples = 0, 0, 0
    progress_bar = tqdm(loader, desc="Training", leave=False)
    for graph_batch, image_batch, target_batch in progress_bar:
        graph_batch, image_batch, target_batch = graph_batch.to(DEVICE), image_batch.to(DEVICE), target_batch.to(DEVICE)
        optimizer.zero_grad()
        out = model(graph_batch, image_batch)
        loss = criterion(out, target_batch)
        if torch.isnan(loss):
            print("!!! Lỗi: Train loss is NaN. Dừng huấn luyện. !!!"); return float('nan'), float('nan')
        loss.backward()
        optimizer.step()

        preds = (torch.sigmoid(out) > 0.5).long()
        total_loss += loss.item() * graph_batch.num_graphs
        total_correct += (preds == target_batch.long()).sum().item()
        total_samples += graph_batch.num_graphs

        progress_bar.set_postfix(loss=loss.item())

    avg_loss = total_loss / total_samples
    avg_acc = total_correct / total_samples
    return avg_loss, avg_acc

def evaluate(model, loader, criterion):
    model.eval()
    total_loss, total_correct, total_samples = 0, 0, 0
    all_preds, all_labels = [], []
    progress_bar = tqdm(loader, desc="Evaluating", leave=False)
    with torch.no_grad():
        for graph_batch, image_batch, target_batch in progress_bar:
            graph_batch, image_batch, target_batch = graph_batch.to(DEVICE), image_batch.to(DEVICE), target_batch.to(DEVICE)
            out = model(graph_batch, image_batch)
            loss = criterion(out, target_batch)

            preds = (torch.sigmoid(out) > 0.5).long()
            total_loss += loss.item() * graph_batch.num_graphs
            total_correct += (preds == target_batch.long()).sum().item()
            total_samples += graph_batch.num_graphs

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(target_batch.cpu().numpy())

    avg_loss = total_loss / total_samples
    avg_acc = total_correct / total_samples
    f1 = f1_score(all_labels, all_preds, zero_division=0)
    return avg_loss, avg_acc, f1

def get_predictions_and_scores(model, loader):
    model.eval()
    all_scores, all_labels = [], []
    with torch.no_grad():
        for graph_batch, image_batch, target_batch in loader:
            graph_batch, image_batch, target_batch = graph_batch.to(DEVICE), image_batch.to(DEVICE), target_batch.to(DEVICE)
            out = model(graph_batch, image_batch)
            scores = torch.sigmoid(out)
            all_scores.extend(scores.cpu().numpy())
            all_labels.extend(target_batch.cpu().numpy())
    return np.array(all_scores).flatten(), np.array(all_labels).flatten()

# ==================================================================
#                       BẮT ĐẦU CHƯƠNG TRÌNH CHÍNH
# ==================================================================
# 1. TẢI VÀ TIỀN XỬ LÝ DỮ LIỆU BẢNG
print("\n[BƯỚC 1/5] Tải và tiền xử lý dữ liệu...")
df = pd.read_excel(EXCEL_FILE_PATH)

### <<< THAY ĐỔI: TÁI CẤU TRÚC LOGIC ĐỂ SỬA LỖI KEYERROR >>> ###
# B1: Làm sạch và lấy nhãn `targets`
df.dropna(subset=[TARGET_COLUMN], inplace=True)
df = df[df[TARGET_COLUMN].isin(['Co', 'Khong'])].copy()
targets = df[TARGET_COLUMN].map({'Co': 1, 'Khong': 0}).values

# B2: Lấy `file_names` một cách an toàn từ dataframe đã được làm sạch
# Điều này đảm bảo `file_names` và `targets` luôn có cùng độ dài và thứ tự
file_names = df['file_name']

# B3: Xác định tất cả các cột không phải là đặc trưng huấn luyện
cols_to_drop = [TARGET_COLUMN, 'file_name', 'hoten', 'ngaysinh', 'sdt', 'id']

# B4: Tạo dataframe chỉ chứa các đặc trưng huấn luyện
features_df_tabular = df.drop(columns=[col for col in cols_to_drop if col in df.columns], errors='ignore')

# B5: Tiền xử lý dataframe đặc trưng này
features_df_tabular.replace(['', ' ', '.', 'khong biet', 'Ko'], np.nan, inplace=True)
categorical_cols = features_df_tabular.select_dtypes(include=['object', 'category']).columns
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
features_df_tabular[categorical_cols] = encoder.fit_transform(features_df_tabular[categorical_cols])
features_df_tabular.fillna(0, inplace=True)
scaler = MinMaxScaler()
features_df_scaled_values = scaler.fit_transform(features_df_tabular)
features_df_scaled = pd.DataFrame(features_df_scaled_values, columns=features_df_tabular.columns, index=features_df_tabular.index)
print(f"Đã xử lý {len(features_df_scaled)} mẫu.")

# 2. XỬ LÝ DỮ LIỆU ẢNH
print("\n[BƯỚC 2/5] Trích xuất đặc trưng từ ảnh...")
safe_file_names = file_names.astype(str).fillna('')
image_features_array = np.array([extract_image_features(os.path.join(IMAGE_DIR, fn)) for fn in tqdm(safe_file_names, desc="Extracting Image Features")])
print(f"Hoàn tất trích xuất đặc trưng cho {len(image_features_array)} mẫu.")

# 3. TẠO DATASET VÀ DATALOADER
print("\n[BƯỚC 3/5] Tạo Dataset và DataLoader...")
indices = list(range(len(features_df_scaled)))
train_indices, test_indices = train_test_split(indices, test_size=0.2, random_state=42, stratify=targets)
train_dataset = OccupationalDiseaseDataset(root=os.path.join(OUTPUT_DIR, 'train'), tabular_df=features_df_scaled.iloc[train_indices].reset_index(drop=True), image_features_array=image_features_array[train_indices], targets=targets[train_indices])
test_dataset = OccupationalDiseaseDataset(root=os.path.join(OUTPUT_DIR, 'test'), tabular_df=features_df_scaled.iloc[test_indices].reset_index(drop=True), image_features_array=image_features_array[test_indices], targets=targets[test_indices])
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)
print("Đã tạo xong DataLoader cho tập train và test.")

# 4. KHỞI TẠO VÀ HUẤN LUYỆN MÔ HÌNH
print("\n[BƯỚC 4/5] Khởi tạo và huấn luyện mô hình...")
model = MultiInputGCN(num_node_features=1).to(DEVICE)
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001, weight_decay=5e-4)
count_class_0 = np.sum(targets == 0); count_class_1 = np.sum(targets == 1)
if count_class_1 == 0:
    weight_for_class_1 = 1.0 # Tránh lỗi chia cho 0
else:
    weight_for_class_1 = count_class_0 / count_class_1
pos_weight = torch.tensor([weight_for_class_1], device=DEVICE)
print(f"Trọng số cho lớp 'Bệnh' (class 1): {pos_weight.item():.2f}")
criterion = BCEWithLogitsLoss(pos_weight=pos_weight)

best_val_f1 = 0
patience, patience_counter = 15, 0
history = {'loss': [], 'acc': [], 'val_loss': [], 'val_acc': [], 'val_f1': []}

for epoch in range(1, 101):
    print(f"\nEpoch {epoch}/{100}")
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion)
    if np.isnan(train_loss): break

    val_loss, val_acc, val_f1 = evaluate(model, test_loader, criterion)

    history['loss'].append(train_loss)
    history['acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    history['val_f1'].append(val_f1)

    print(f"--> Epoch {epoch} Summary: loss: {train_loss:.4f} - acc: {train_acc:.4f} - val_loss: {val_loss:.4f} - val_acc: {val_acc:.4f} - val_f1: {val_f1:.4f}")

    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        torch.save(model.state_dict(), os.path.join(OUTPUT_DIR, MODEL_FILE_NAME))
        print(f"*** Val F1-Score improved to {best_val_f1:.4f}. Saving model... ***")
        patience_counter = 0
    else:
        patience_counter += 1
        print(f"Val F1-Score did not improve. Patience: {patience_counter}/{patience}")

    if patience_counter >= patience:
        print(f"Early stopping at epoch {epoch}.")
        break

# 5. LƯU CÁC FILE CẦN THIẾT
print("\n[BƯỚC 5/5] Lưu các tệp tiền xử lý...")
with open(os.path.join(OUTPUT_DIR, COLUMNS_FILE_NAME), 'wb') as f: pickle.dump(features_df_scaled.columns.tolist(), f)
with open(os.path.join(OUTPUT_DIR, ENCODER_FILE_NAME), 'wb') as f: pickle.dump(encoder, f)
with open(os.path.join(OUTPUT_DIR, SCALER_FILE_NAME), 'wb') as f: pickle.dump(scaler, f)
print(f"QUÁ TRÌNH HUẤN LUYỆN HOÀN TẤT! Model và tệp đã được lưu vào: {OUTPUT_DIR}")

# ==================================================================
#                BƯỚC 6: PHÂN TÍCH VÀ VẼ BIỂU ĐỒ
# ==================================================================
print("\n[BƯỚC 6/6] Tạo các biểu đồ phân tích mô hình...")
model.load_state_dict(torch.load(os.path.join(OUTPUT_DIR, MODEL_FILE_NAME)))

# ---- 1. Biểu đồ Loss (Train vs. Validation) ----
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 5))
plt.plot(history['loss'], label='Train Loss', color='blue')
plt.plot(history['val_loss'], label='Validation Loss', color='orange')
plt.title('Training & Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# ---- 2. Biểu đồ Accuracy (Train vs. Validation) ----
plt.figure(figsize=(12, 5))
plt.plot(history['acc'], label='Train Accuracy', color='blue')
plt.plot(history['val_acc'], label='Validation Accuracy', color='orange')
plt.title('Training & Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# ---- 3. Phân tích Chi tiết trên Tập Test ----
y_scores, y_true = get_predictions_and_scores(model, test_loader)
y_pred = (y_scores > 0.5).astype(int)

# Báo cáo Phân loại
print("\nClassification Report trên tập Test:")
print(classification_report(y_true, y_pred, target_names=['Không Bệnh (0)', 'Bệnh (1)']))

# Ma trận nhầm lẫn
cm = confusion_matrix(y_true, y_pred); plt.figure(figsize=(8, 6)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Không Bệnh', 'Bệnh'], yticklabels=['Không Bệnh', 'Bệnh']); plt.xlabel('Dự đoán'); plt.ylabel('Thực tế'); plt.title('Ma trận nhầm lẫn'); plt.show()

# ---- 4. Biểu đồ ROC và chỉ số AUC ----
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
auc_score = roc_auc_score(y_true, y_scores)
plt.figure(figsize=(8, 8)); plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc_score:.2f})'); plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--'); plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05]); plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate'); plt.title('Receiver Operating Characteristic (ROC) Curve'); plt.legend(loc="lower right"); plt.show()
```

```python
# -*- coding: utf-8 -*-
"""
Mã nguồn hoàn chỉnh để huấn luyện mô hình bệnh nghề nghiệp
kết hợp dữ liệu bảng và dữ liệu hình ảnh.
*** PHIÊN BẢN KẾT HỢP: Sử dụng logic tiền xử lý từ code cũ (Stellargraph)
                     và kiến trúc mô hình mới (PyTorch Geometric) ***
"""

# BƯỚC 1: CÀI ĐẶT CÁC THƯ VIỆN CẦN THIẾT
# -------------------------------------------------


import os
import gc
import pickle
import numpy as np
import pandas as pd
from PIL import Image
from tqdm.notebook import tqdm

import torch
import torch.nn.functional as F
from torch.nn import Linear, BatchNorm1d, Module, Sequential, ReLU, Dropout, BCEWithLogitsLoss
from torch_geometric.data import Data, Dataset, DataLoader
from torch_geometric.nn import GCNConv, global_mean_pool

import torchvision.transforms as T
from torchvision.models import mobilenet_v2, MobileNet_V2_Weights
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score, f1_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns


# BƯỚC 2: THIẾT LẬP MÔI TRƯỜNG VÀ ĐƯỜNG DẪN
# -------------------------------------------------
from google.colab import drive
drive.mount('/content/drive')

BASE_DIR = "/content/"
EXCEL_FILE_PATH = os.path.join(BASE_DIR, "Main_data_fixed_Not_Encode_Mapping_New.xlsx")
IMAGE_DIR = os.path.join(BASE_DIR, "sillicosis")
OUTPUT_DIR = os.path.join(BASE_DIR, "output_model_pytorch_legacy_logic")

os.makedirs(OUTPUT_DIR, exist_ok=True)
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Sử dụng thiết bị: {DEVICE}")

# --- Các hằng số ---
IMAGE_VECTOR_DIM = 1280
TARGET_COLUMN = 'bnn'
MODEL_FILE_NAME = "occupational_disease_pyg.pth"
# Các file lưu trữ logic tiền xử lý cũ
COLUMNS_FILE_NAME = "LEGACY_COLUMNS.pkl"
MEAN_VALUES_FILE_NAME = "LEGACY_MEAN_VALUES.pkl"
MAX_VALUES_FILE_NAME = "LEGACY_MAX_VALUES.pkl"


# BƯỚC 3: TRÍCH XUẤT ĐẶC TRƯNG ẢNH
# -------------------------------------------------
print("Bắt đầu tải mô hình trích xuất đặc trưng ảnh...")
weights = MobileNet_V2_Weights.IMAGENET1K_V2
image_feature_extractor = mobilenet_v2(weights=weights)
image_feature_extractor.classifier = torch.nn.Identity()
image_feature_extractor.to(DEVICE)
image_feature_extractor.eval()
image_transforms = weights.transforms()

def extract_image_features(image_path):
    if not image_path or not os.path.exists(image_path):
        return np.zeros(IMAGE_VECTOR_DIM)
    try:
        img = Image.open(image_path).convert('RGB')
        img_tensor = image_transforms(img).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            features = image_feature_extractor(img_tensor)
        return features.cpu().numpy().flatten()
    except Exception as e:
        return np.zeros(IMAGE_VECTOR_DIM)


### <<< THAY ĐỔI 1: TÍCH HỢP LOGIC TIỀN XỬ LÝ TỪ CODE CŨ >>> ###
# ==================================================================
#         BƯỚC 4: CÁC HÀM TIỀN XỬ LÝ THEO LOGIC CŨ (LEGACY)
# ==================================================================

def preprocess_data_legacy_style(X):
    """Hàm này chứa toàn bộ logic mã hóa từ hàm `encoded()` trong code cũ."""
    print("Áp dụng logic mã hóa thủ công (legacy)...")

    # Chuyển đổi các giá trị chung
    X.replace(['khong','co','ko','khong hut thuoc','co hut thuoc'], [0,1,0,0,1], inplace = True)
    X.replace(['', ' ', '.','khong biet'], np.nan, inplace = True)

    # Chuyển đổi các cột cụ thể
    # Sử dụng try-except để bỏ qua nếu cột không tồn tại
    try:
        X['B13'].replace(['1.0','4.0','145.0','12345.0','1234.0','134.0','1245.0','13.0','14.0','1345.0','3.0','34.0','124.0','45.0','5.0','15.0','12.0','24.0','2345.0','345.0','123.0','245.0','2.0','125.0','7.0','234.0','23.0','1435.0','25.0','1453.0','35.0','135.0','13245.0','235.0','1235.0','10.0','156.0'],
                            [1,4,10,15,10,8,12,4,5,13,3,7,7,9,5,6,3,6,14,12,6,11,2,8,7,9,5,13,7,13,8,9,15,10,11,1,12], inplace = True)
        X['C7'].replace(['1.0','123.0','5.0','13.0','3.0','12.0','2.0','23.0','4.0','14.0','134.0','124.0','1234.0','24.0','34.0'],
                                [1,6,5,4,3,3,2,5,4,5,8,7,10,6,7], inplace = True)
        X['C13'].replace(['2.0','5.0','12345.0','2345.0','','1.0','245.0','24.0','15.0','25.0','4.0','45.0','1245.0','125.0','12.0','2456.0','6.0','234.0','1235.0','345.0','145.0','34.0','1234.0','235.0','3456.0','23.0','35.0','3.0','246.0','124.0','456.0','1345.0','123456.0','123.0','12456.0','23456.0','135.0','56.0','156.0','14.0','13.0','356.0'],
                                [2,5,15,14,np.nan,1,11,6,6,7,4,9,12,8,3,17,6,9,11,12,10,7,10,10,18,5,8,3,12,7,15,13,21,6,18,20,9,11,12,5,4,14], inplace = True)
        X['A6'].replace(['Da ket hon va dang chung song','Chua ket hon','Li di','Goa','Li than'], [2,0,1,0,1], inplace = True)
        X['tsnoi'].replace(['khong', r'^\s*(?!0$).+\s*$'], [0, 1], regex=True, inplace = True)
        X['tsngoai'].replace(['khong', r'^\s*(?!0$).+\s*$'],  [0, 1], regex=True, inplace = True)
        X['rungthan'].replace(['giam', 'tang'], [-1, 1], regex=True, inplace = True)
        X['tsho'].replace(['khac','lien tuc','tung con'], [0,2,1], inplace = True)
        X['vitridau'].replace(['hai ben nguc','nguc phai','dau sau xuong uc','nguc trai'], [3,1,4,2], inplace = True)
        X['longnguc'].replace(['can doi','khong can doi'], [1,0], inplace = True)
        X['khoangls'].replace(['Binh thuong', 'Gian rong', 'Hep'], [0,1,-1], inplace = True)
        X['go'].replace(['go duc'], [1], inplace = True)
        X['riraopn'].replace(['giam','khac'], [-1,1], inplace = True)
        X['B7'].replace(['hut hang nay','hut 1 tuan 1 lan','hiem khi hut','hut 1 thang 1 lan'],[2,1,-1,0], inplace = True)
    except KeyError as e:
        print(f"Cảnh báo: Cột {e} không tồn tại để thay thế giá trị.")

    # Chuyển đổi các giá trị chung khác
    X.replace(['binh thuong','tot','rat tot','khong tot'],[0,1,2,-1], inplace = True)
    X.replace(['khong thoai mai','thoai mai','rat khong thoai mai','rat thoai mai'],[-1,1,-2,2], inplace = True)
    X.replace(['nhieu','trung binh','rat nhieu','it','rat it'],[1,0,2,-1,-2], inplace = True)
    X.replace(['nguy hiem','rat nguy hiem','khong nguy hiem'],[1,2,0], inplace = True)
    X.replace(['de mac','rat de mac','khong the mac'],[0,1,-1], inplace = True)
    X.replace(['rat thuong xuyen','thuong xuyen','thinh thoang','hiem khi','khong su dung'],[2,1,0,-1,-2], inplace = True)
    X.replace(['ban ngay','ca ngay va dem','ban dem'],[0,2,1], inplace = True)

    # Chuyển đổi tất cả các cột sang dạng số, nếu không được thì bỏ qua
    for col in X.columns:
        X[col] = pd.to_numeric(X[col], errors='coerce')

    return X

def normalize_data_legacy_style(df, mean_vals, max_vals):
    """Hàm chuẩn hóa dữ liệu theo logic cũ."""
    print("Áp dụng logic chuẩn hóa (legacy)...")
    df_normalized = df.copy()
    for column in df_normalized.columns:
        if column in mean_vals and column in max_vals:
            denominator = max_vals[column] - mean_vals[column]
            if denominator != 0:
                df_normalized[column] = (df_normalized[column] - mean_vals[column]) / denominator
    return df_normalized

# ==================================================================
#         KẾT THÚC PHẦN LOGIC TIỀN XỬ LÝ CŨ
# ==================================================================


# BƯỚC 5: XÂY DỰNG DATASET TÙY CHỈNH
# -------------------------------------------------
class OccupationalDiseaseDataset(Dataset):
    def __init__(self, root, tabular_df, image_features_array, targets):
        super(OccupationalDiseaseDataset, self).__init__(root)
        self.tabular_df, self.image_features_array, self.targets = tabular_df, image_features_array, targets
        self.columns = tabular_df.columns.tolist()
        self._build_graphs()

    def len(self):
        return len(self.graphs)

    def get(self, idx):
        graph_data = self.graphs[idx]
        image_features = torch.tensor(self.image_features_array[idx], dtype=torch.float)
        target = torch.tensor([self.targets[idx]], dtype=torch.float)
        return graph_data, image_features, target

    def _build_graphs(self):
        self.graphs = []
        relation_ship_groups = [['cviec1','tgian1'],['cviec2','tgian2'],['tuoinghe','A9a','A9b','nampx'],['hutthuoc','slthuoc'],['tiensuhh','benhhh'],['ho','tdho','tsho']]
        for _, row in self.tabular_df.iterrows():
            # Chỉ tạo node cho các đặc trưng có giá trị (không phải NaN)
            valid_features = row.dropna()
            node_features = torch.tensor(valid_features.values, dtype=torch.float).view(-1, 1)

            # Nếu không có đặc trưng nào hợp lệ, tạo một graph rỗng
            if node_features.shape[0] == 0:
                self.graphs.append(Data(x=torch.empty(0, 1), edge_index=torch.empty(2, 0, dtype=torch.long)))
                continue

            # Xây dựng ma trận kề
            source_nodes, target_nodes = [], []
            num_nodes = len(valid_features)
            original_indices = {col: i for i, col in enumerate(valid_features.index)}

            # Kết nối các node với nhau
            for i in range(num_nodes):
                for j in range(i + 1, num_nodes):
                    source_nodes.extend([i, j]); target_nodes.extend([j, i])

            # Thêm các cạnh quan hệ cụ thể
            for group in relation_ship_groups:
                # Lấy ra các cột trong nhóm có tồn tại trong các đặc trưng hợp lệ của mẫu này
                valid_group_cols = [col for col in group if col in valid_features.index]
                if len(valid_group_cols) > 1:
                    # Chuyển tên cột thành chỉ số node tương ứng
                    node_indices_in_graph = [original_indices[col] for col in valid_group_cols]
                    first_node = node_indices_in_graph[0]
                    for other_node in node_indices_in_graph[1:]:
                        source_nodes.extend([first_node, other_node]); target_nodes.extend([other_node, first_node])

            edge_index = torch.tensor([source_nodes, target_nodes], dtype=torch.long)
            self.graphs.append(Data(x=node_features, edge_index=edge_index))


# BƯỚC 6: ĐỊNH NGHĨA KIẾN TRÚC MÔ HÌNH
# -------------------------------------------------
class MultiInputGCN(Module):
    def __init__(self, num_node_features, graph_embedding_dim=64, image_feature_dim=IMAGE_VECTOR_DIM):
        super(MultiInputGCN, self).__init__()
        self.gcn1 = GCNConv(num_node_features, 128)
        self.gcn2 = GCNConv(128, graph_embedding_dim)
        self.image_mlp = Sequential(Linear(image_feature_dim, 256), ReLU(), Dropout(0.5), Linear(256, graph_embedding_dim))
        self.classifier = Sequential(Linear(graph_embedding_dim * 2, 128), ReLU(), BatchNorm1d(128), Dropout(0.5), Linear(128, 1))

    def forward(self, graph_data, image_features):
        x, edge_index, batch = graph_data.x, graph_data.edge_index, graph_data.batch
        x = F.relu(self.gcn1(x, edge_index))
        x = self.gcn2(x, edge_index)
        graph_embedding = global_mean_pool(x, batch)
        image_embedding = self.image_mlp(image_features)
        combined_embedding = torch.cat([graph_embedding, image_embedding], dim=1)
        return self.classifier(combined_embedding)


# BƯỚC 7: TIẾN TRÌNH HUẤN LUYỆN VÀ ĐÁNH GIÁ
# -------------------------------------------------
def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss, total_correct, total_samples = 0, 0, 0
    progress_bar = tqdm(loader, desc="Training", leave=False)
    for graph_batch, image_batch, target_batch in progress_bar:
        graph_batch, image_batch, target_batch = graph_batch.to(DEVICE), image_batch.to(DEVICE), target_batch.to(DEVICE)
        optimizer.zero_grad()
        out = model(graph_batch, image_batch)
        loss = criterion(out, target_batch)
        if torch.isnan(loss):
            print("!!! Lỗi: Train loss is NaN. Dừng huấn luyện. !!!"); return float('nan'), float('nan')
        loss.backward()
        optimizer.step()
        preds = (torch.sigmoid(out) > 0.5).long()
        total_loss += loss.item() * graph_batch.num_graphs
        total_correct += (preds == target_batch.long()).sum().item()
        total_samples += graph_batch.num_graphs
        progress_bar.set_postfix(loss=loss.item())
    avg_loss = total_loss / total_samples
    avg_acc = total_correct / total_samples
    return avg_loss, avg_acc

def evaluate(model, loader, criterion):
    model.eval()
    total_loss, total_correct, total_samples = 0, 0, 0
    all_preds, all_labels = [], []
    progress_bar = tqdm(loader, desc="Evaluating", leave=False)
    with torch.no_grad():
        for graph_batch, image_batch, target_batch in progress_bar:
            graph_batch, image_batch, target_batch = graph_batch.to(DEVICE), image_batch.to(DEVICE), target_batch.to(DEVICE)
            out = model(graph_batch, image_batch)
            loss = criterion(out, target_batch)
            preds = (torch.sigmoid(out) > 0.5).long()
            total_loss += loss.item() * graph_batch.num_graphs
            total_correct += (preds == target_batch.long()).sum().item()
            total_samples += graph_batch.num_graphs
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(target_batch.cpu().numpy())
    avg_loss = total_loss / total_samples
    avg_acc = total_correct / total_samples
    f1 = f1_score(all_labels, all_preds, zero_division=0)
    return avg_loss, avg_acc, f1

def get_predictions_and_scores(model, loader):
    model.eval()
    all_scores, all_labels = [], []
    with torch.no_grad():
        for graph_batch, image_batch, target_batch in loader:
            graph_batch, image_batch, target_batch = graph_batch.to(DEVICE), image_batch.to(DEVICE), target_batch.to(DEVICE)
            out = model(graph_batch, image_batch)
            scores = torch.sigmoid(out)
            all_scores.extend(scores.cpu().numpy())
            all_labels.extend(target_batch.cpu().numpy())
    return np.array(all_scores).flatten(), np.array(all_labels).flatten()


# ==================================================================
#                       BẮT ĐẦU CHƯƠNG TRÌNH CHÍNH
# ==================================================================
# 1. TẢI VÀ TIỀN XỬ LÝ DỮ LIỆU BẢNG
print("\n[BƯỚC 1/5] Tải và tiền xử lý dữ liệu...")
df = pd.read_excel(EXCEL_FILE_PATH)
df.dropna(subset=[TARGET_COLUMN], inplace=True)
df = df[df[TARGET_COLUMN].isin(['Co', 'Khong'])].copy()
targets = df[TARGET_COLUMN].map({'Co': 1, 'Khong': 0}).values
file_names = df['file_name']
cols_to_drop = [TARGET_COLUMN, 'hoten', 'ngaysinh', 'sdt', 'id']
features_df = df.drop(columns=[col for col in cols_to_drop if col in df.columns], errors='ignore')

### <<< THAY ĐỔI 2: ÁP DỤNG CÁC HÀM TIỀN XỬ LÝ THEO LOGIC CŨ >>> ###
# B1: Mã hóa thủ công
features_df_encoded = preprocess_data_legacy_style(features_df.copy())

# B2: Tính toán và lưu trữ các giá trị chuẩn hóa
# Chỉ tính trên tập train để tránh rò rỉ dữ liệu từ tập test
temp_train_indices, _ = train_test_split(range(len(features_df_encoded)), test_size=0.2, random_state=42, stratify=targets)
train_df_for_scaling = features_df_encoded.iloc[temp_train_indices]

mean_values = train_df_for_scaling.mean(numeric_only=True)
max_values = train_df_for_scaling.max(numeric_only=True)

with open(os.path.join(OUTPUT_DIR, MEAN_VALUES_FILE_NAME), 'wb') as f: pickle.dump(mean_values, f)
with open(os.path.join(OUTPUT_DIR, MAX_VALUES_FILE_NAME), 'wb') as f: pickle.dump(max_values, f)
with open(os.path.join(OUTPUT_DIR, COLUMNS_FILE_NAME), 'wb') as f: pickle.dump(features_df_encoded.columns.tolist(), f)

# B3: Chuẩn hóa toàn bộ dữ liệu bằng các giá trị đã tính
features_df_scaled = normalize_data_legacy_style(features_df_encoded, mean_values, max_values)
# Lấp đầy các giá trị NaN có thể phát sinh sau khi chuẩn hóa (ví dụ: chia cho 0)
features_df_scaled.fillna(0, inplace=True)
print(f"Đã xử lý {len(features_df_scaled)} mẫu theo logic cũ.")

# 2. XỬ LÝ DỮ LIỆU ẢNH
print("\n[BƯỚC 2/5] Trích xuất đặc trưng từ ảnh...")
safe_file_names = file_names.astype(str).fillna('')
image_features_array = np.array([extract_image_features(os.path.join(IMAGE_DIR, fn)) for fn in tqdm(safe_file_names, desc="Extracting Image Features")])
print(f"Hoàn tất trích xuất đặc trưng cho {len(image_features_array)} mẫu.")

# 3. TẠO DATASET VÀ DATALOADER
print("\n[BƯỚC 3/5] Tạo Dataset và DataLoader...")
indices = list(range(len(features_df_scaled)))
train_indices, test_indices = train_test_split(indices, test_size=0.2, random_state=42, stratify=targets)
train_dataset = OccupationalDiseaseDataset(root=os.path.join(OUTPUT_DIR, 'train'), tabular_df=features_df_scaled.iloc[train_indices].reset_index(drop=True), image_features_array=image_features_array[train_indices], targets=targets[train_indices])
test_dataset = OccupationalDiseaseDataset(root=os.path.join(OUTPUT_DIR, 'test'), tabular_df=features_df_scaled.iloc[test_indices].reset_index(drop=True), image_features_array=image_features_array[test_indices], targets=targets[test_indices])
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)
print("Đã tạo xong DataLoader cho tập train và test.")

# 4. KHỞI TẠO VÀ HUẤN LUYỆN MÔ HÌNH
print("\n[BƯỚC 4/5] Khởi tạo và huấn luyện mô hình...")
model = MultiInputGCN(num_node_features=1).to(DEVICE)
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001, weight_decay=5e-4)
count_class_0 = np.sum(targets == 0); count_class_1 = np.sum(targets == 1)
if count_class_1 == 0: weight_for_class_1 = 1.0
else: weight_for_class_1 = count_class_0 / count_class_1
pos_weight = torch.tensor([weight_for_class_1], device=DEVICE)
print(f"Trọng số cho lớp 'Bệnh' (class 1): {pos_weight.item():.2f}")
criterion = BCEWithLogitsLoss(pos_weight=pos_weight)

best_val_f1 = 0
patience, patience_counter = 15, 0
history = {'loss': [], 'acc': [], 'val_loss': [], 'val_acc': [], 'val_f1': []}

for epoch in range(1, 101):
    print(f"\nEpoch {epoch}/{100}")
    train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion)
    if np.isnan(train_loss): break
    val_loss, val_acc, val_f1 = evaluate(model, test_loader, criterion)
    history['loss'].append(train_loss); history['acc'].append(train_acc)
    history['val_loss'].append(val_loss); history['val_acc'].append(val_acc); history['val_f1'].append(val_f1)
    print(f"--> Epoch {epoch} Summary: loss: {train_loss:.4f} - acc: {train_acc:.4f} - val_loss: {val_loss:.4f} - val_acc: {val_acc:.4f} - val_f1: {val_f1:.4f}")
    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        torch.save(model.state_dict(), os.path.join(OUTPUT_DIR, MODEL_FILE_NAME))
        print(f"*** Val F1-Score improved to {best_val_f1:.4f}. Saving model... ***")
        patience_counter = 0
    else:
        patience_counter += 1
        print(f"Val F1-Score did not improve. Patience: {patience_counter}/{patience}")
    if patience_counter >= patience:
        print(f"Early stopping at epoch {epoch}.")
        break

# 5. LƯU CÁC FILE CẦN THIẾT
# (Lưu ý: các file này giờ chứa logic của code cũ)
print("\n[BƯỚC 5/5] Lưu các tệp tiền xử lý...")
# Đã lưu ở bước 1, không cần lưu lại ở đây.
print(f"QUÁ TRÌNH HUẤN LUYỆN HOÀN TẤT! Model và tệp đã được lưu vào: {OUTPUT_DIR}")


# ==================================================================
#                BƯỚC 6: PHÂN TÍCH VÀ VẼ BIỂU ĐỒ
# ==================================================================
print("\n[BƯỚC 6/6] Tạo các biểu đồ phân tích mô hình...")
model.load_state_dict(torch.load(os.path.join(OUTPUT_DIR, MODEL_FILE_NAME)))

plt.style.use('seaborn-v0_8-whitegrid')
# ---- Biểu đồ Loss ----
plt.figure(figsize=(12, 5)); plt.plot(history['loss'], label='Train Loss', color='blue'); plt.plot(history['val_loss'], label='Validation Loss', color='orange'); plt.title('Training & Validation Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.show()
# ---- Biểu đồ Accuracy ----
plt.figure(figsize=(12, 5)); plt.plot(history['acc'], label='Train Accuracy', color='blue'); plt.plot(history['val_acc'], label='Validation Accuracy', color='orange'); plt.title('Training & Validation Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.show()

y_scores, y_true = get_predictions_and_scores(model, test_loader)
y_pred = (y_scores > 0.5).astype(int)
print("\nClassification Report trên tập Test:"); print(classification_report(y_true, y_pred, target_names=['Không Bệnh (0)', 'Bệnh (1)']))
cm = confusion_matrix(y_true, y_pred); plt.figure(figsize=(8, 6)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Không Bệnh', 'Bệnh'], yticklabels=['Không Bệnh', 'Bệnh']); plt.xlabel('Dự đoán'); plt.ylabel('Thực tế'); plt.title('Ma trận nhầm lẫn'); plt.show()
fpr, tpr, _ = roc_curve(y_true, y_scores); auc_score = roc_auc_score(y_true, y_scores)
plt.figure(figsize=(8, 8)); plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc_score:.2f})'); plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--'); plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05]); plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate'); plt.title('Receiver Operating Characteristic (ROC) Curve'); plt.legend(loc="lower right"); plt.show()
```
